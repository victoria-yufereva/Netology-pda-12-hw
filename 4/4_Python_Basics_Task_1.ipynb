Домашнее задание к лекции "Функции" Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется); в коде отсутствуют глобальные переменные (за исключением documents и directories); пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").



```python
# перечень всех документов
documents = [ 
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]
# перечень полок, на которых хранятся документы (если документ есть в)
directories = { 
    '1': ['2207 876234', '11-2'], 
    '2': ['10006'], 
    '3': [] 
} 


def owner_by_number():
    """"
    Пользователь может узнать владельца документа по его номеру
    """
    match = 0
    user_input = input('Введите номер документа: ')
    for element in documents:
        if user_input == element['number']:
            match = 1
            owner = element['name']
            result_yes = f'Владелец документа: {owner}'
        else:
            result_no = 'Документ не найден в базе'
    if match != 0:
        print(result_yes)
    else:
        print(result_no)

# owner_by_number()

def shelf_by_number():
    """
    Пользователь может по номеру документа узнать на какой полке он хранится
    """
    match_doc = 0
    user_input = input('Введите номер документа: ')
    for element in documents:
        if user_input == element['number']:
            match_doc = 1
            for shelf, number_list in directories.items():
                for num in number_list:
                    if num == element['number']:
                        result_shelf = f'Документ хранится на полке:{shelf}'
    if match_doc != 0:
        print(result_shelf)
    else:
        print('Документ не найден в базе')

# shelf_by_number()

def all_info():
    """
    Пользователь может увидеть полную информацию по всем документам
    """
    for person in documents:
        for element in person:
            number = person['number']
            doc_type = person['type']
            owner_name = person['name']
            result = f'№: {number}, тип: {doc_type}, владелец: {owner_name},'
            for shelf, number_list in directories.items():
                for number in number_list:
                    res_shelf = f'полка хранения:{shelf}'
        print(result, res_shelf)
        
# all_info()

def add_new_shelf():
    """
    Пользователь может добавить новую полку
    """
    user_input = input('Введите номер полки, которую хотите добавить: ')
    shelves_list = list(directories.keys())
    if user_input in directories:
        print('Такая полка уже существует. Текущий перечень полок: ', end='')
        print(*shelves_list, sep= ', ')
    else:
        directories[user_input] = []
        shelves_list = list(directories.keys())
        print('Полка добавлена. Текущий перечень полок: ', end='')
        print(*shelves_list, sep= ', ')
        
# add_new_shelf()

def delete_empty_shelf():
    """
    Пользователь может удалить существующую полку из данных (только если она пустая)
    """
    user_input = input('Введите номер полки, которую хотите удалить: ')
    if user_input in directories and len(directories[user_input]) == 0:
        del(directories[user_input])
        print('Полка удалена. Текущий перечень полок: ', end='')
        shelves_list = list(directories.keys())
        print(*shelves_list, sep= ', ')
    elif user_input in directories and len(directories[user_input]) != 0:
        print(directories[user_input])
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', end='')
        shelves_list = list(directories.keys())
        print(*shelves_list, sep= ', ')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        shelves_list = list(directories.keys())
        print(*shelves_list, sep= ', ')
        
# delete_empty_shelf()

def main(documents, directories):
    """
    Пользователь программы:
    - по команде "p" может узнать владельца документа по его номеру
    - по команде "s" может по номеру документа узнать на какой полке он хранится
    - по команде "l" может увидеть полную информацию по всем документам
    - по команде "as" может добавить новую полку
    - по команде "ds" может удалить существующую полку из данных (только если она пустая)
    - может завершить программу вводом команды "q"
    
    """
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            owner_by_number()
        elif user_input == 's':
            shelf_by_number()
        elif user_input == 'l':
            all_info()
        elif user_input == 'as':
            add_new_shelf()
        elif user_input == 'ds':
            delete_empty_shelf()
        elif user_input == 'q':
            print('Работа программы завершена')
            break  
            
main(documents, directories)
```
