### Работаем с заменами пустых значений

```python
data.rec_temp.fillna(data.rec_temp.median(), inplace=True)
data.pulse.fillna(data.pulse.median(), inplace=True)
data.total_protein.fillna(data.total_protein.median(), inplace=True)
data.describe()
```

```python
data.surgery.fillna(data.surgery.mode()[0], inplace=True)
data.outcome.fillna(data.outcome.mode()[0], inplace=True)
data.temp_of_extremities.fillna(data.temp_of_extremities.mode()[0], inplace=True)
```

### Работаем с выбросами количественных показателей
### Находим выбросы через межкваритльный размах

```python
# считаем выбросы показателя rec_temp после замены пропусков значений (у нас было 60 пропущенных из 300):
# видим, что у нас 280 строк осталось после удаления выбросов:
q1 = data.rec_temp.quantile(0.25)
q3 = data.rec_temp.quantile(0.75)
iqr = q3 - q1
print('iqr (rec_temp) :', iqr)
lower_bound = q1 - (1.5 * iqr)
print('lower_bound :', lower_bound)
upper_bound = q3 + (1.5 * iqr)
print('upper_bound :', upper_bound)
data_2 = data[data.rec_temp.between(lower_bound, upper_bound, inclusive=True)].sort_values('rec_temp')
data_2
```

### Мы нашли верхние выбросы в rec_tem - начиная от 39.4 С до максимального значения - 40.8 С
### Также мы нашли нижние выбросы от 35.4 С до минимального значения - 37.0 С
##### Чтобы оценить их влияние на базовые метрики проведем расчеты
Как видим далее самое чувствительное к выбросам - среднеарифмерическое (старое значение 38.174333 и новое значение 38.166071) поменялось незначительно. Решение - не удалять эти выбросы, поскольку они с большой вероятностью являются исключительными наблюдениями в данных, с меньшей уверенностью можно говорить о погрешности измерительных приборов

```python
# считаем выбросы показателя pulse после замены пропусков значений (до этого у нас было 24 пропущенных из 300):
# видим, что у нас 295 строк осталось после удаления выбросов:
q1 = data.pulse.quantile(0.25)
q3 = data.pulse.quantile(0.75)
iqr = q3 - q1
print('iqr (pulse) :', iqr)
lower_bound = q1 - (1.5 * iqr) 
print('lower_bound :', lower_bound)
upper_bound = q3 + (1.5 * iqr)
print('upper_bound  :', upper_bound)
data = data[data.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
data
```
